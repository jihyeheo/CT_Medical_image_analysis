```python
import numpy as np
def make_noise(std):
    height, width = gray.shape
    img_noise = np.zeros((height, width), dtype=np.float)
    for i in range(height): 
        for a in range(width): 
            make_noise = np.random.normal() 
            set_noise = std * make_noise 
            img_noise[i][a] = gray[i][a] + set_noise
    return img_noise
```
